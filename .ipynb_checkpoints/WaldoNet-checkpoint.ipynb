```python
import numpy as np
import pandas as pd
import cv2
from sklearn.model_selection import train_test_split
from glob import glob
import matplotlib.pyplot as plt
from keras.layers import MaxPool2D, Conv2D, Dense, Dropout, Flatten, Input
from keras.models import Model, load_model
from keras.optimizers import Adam
```

```python
X =  np.asarray([cv2.cvtColor(cv2.imread(x), cv2.COLOR_BGR2RGB) for x in glob("data/256/waldo/*")] +
                [cv2.cvtColor(cv2.imread(x), cv2.COLOR_BGR2RGB) for x in glob("data/256/notwaldo/*")])
Y = np.asarray([1 for x in glob("data/256/waldo/*")] + [0 for x in glob("data/256/notwaldo/*")])
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=25)
x_train.shape, x_test.shape
```

```python
fig=plt.figure(figsize=(8, 8))
columns = 4
rows = 5
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(x_train[i])
```

```python
plt.imshow(x_train[0])
```

```python
inputs = Input(shape=x_train.shape[1:])
outputs = Conv2D(8, (3, 3), activation='relu', input_shape=x_train.shape[1:])(inputs)
outputs = MaxPool2D(pool_size=(2,2))(outputs)
outputs = Flatten()(outputs)
outputs = Dense(1, activation='sigmoid')(outputs)
model = Model(inputs=inputs, outputs=outputs)
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, batch_size=16)
```

```python
print("Loss: %f, Accuracy: %f" % tuple(model.evaluate(x_test, y_test, verbose=0)))
```
